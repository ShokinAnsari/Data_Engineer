```python
# Operations on dataframe 
weather_df['WindSpeed_kmph'] + 3 # +,-,div,mul
```

```python
weather_df['new_Column'] = weather_df['WindSpeed_kmph'] + weather_df['Temperature_C']
weather_df['new_Column']
```

```python
# Adding\Updating cols
updated_col = weather_df['Temperature_C'] * 100
weather_df['Temperature_C']=updated_col
weather_df.head()
```

```python
#.apply(function_name) 
def Solve(num):
    return num + 2

weather_df['Temperature_C'].apply(Solve)
```

```python
#Rename -> dictionary se ho raha hai
weather_df.rename(columns={'Temperature_C':'Temp','Condition':'mausam'},inplace=True) # inplace=True mtlb current df pe apply karo
weather_df.head()
```

```python
weather_df.head() = weather_df.drop(columns=['new_Column']) # delete columns from df
weather_df.head()
```
