```python
import json
```

```python
data={
    'name':{
    'age':'23',
    'roll':'45'
    }
}
```

```python
with open('data_file.json','w') as write_file: #Writing JSON file
    json.dump(data,write_file) #note: dump() takes two positional arguments
```

```python
type(data)
```

```python
#Reading JSON file 
with open('data_file.json','r') as read_file:
    data = json.load(read_file)
```
